## Download sample data

For demonstration purpose, download some sample data from [kuzudb.com](https://kuzudb.com) and create a local database with it.

```python
import requests

urls = [
    "https://raw.githubusercontent.com/ladybugdb/ladybug/refs/heads/master/dataset/demo-db/csv/city.csv",
    "https://raw.githubusercontent.com/ladybugdb/ladybug/refs/heads/master/dataset/demo-db/csv/user.csv",
    "https://raw.githubusercontent.com/ladybugdb/ladybug/refs/heads/master/dataset/demo-db/csv/follows.csv",
    "https://raw.githubusercontent.com/ladybugdb/ladybug/refs/heads/master/dataset/demo-db/csv/lives-in.csv"
]

for url in urls:
    filename = url.split("/")[-1]
    response = requests.get(url)
    with open(filename, 'wb') as f:
        f.write(response.content)
    print(f"Downloaded {filename}")
```

## Create sample database

Use the dowloaded CSV files to create a sample database.

```python
import real_ladybug as lb

# Create an empty on-disk database and connect to it
db = lb.Database("demo-db2.lbug")
conn = lb.Connection(db)

# Create schema
conn.execute("CREATE NODE TABLE User(name STRING PRIMARY KEY, age INT64)")
conn.execute("CREATE NODE TABLE City(name STRING PRIMARY KEY, population INT64)")
conn.execute("CREATE REL TABLE Follows(FROM User TO User, since INT64)")
conn.execute("CREATE REL TABLE LivesIn(FROM User TO City)")

# Insert data
conn.execute('COPY User FROM "./user.csv"')
conn.execute('COPY City FROM "./city.csv"')
conn.execute('COPY Follows FROM "./follows.csv"')
conn.execute('COPY LivesIn FROM "./lives-in.csv"')
```

# Default, no-config visualization

Easily visualize the result of a query.

```python
from yfiles_jupyter_graphs_for_kuzu import KuzuGraphWidget
g = KuzuGraphWidget(conn)
g.show_cypher('MATCH (a:User)-[f:Follows]->(b:User) RETURN a, f, b')
```
